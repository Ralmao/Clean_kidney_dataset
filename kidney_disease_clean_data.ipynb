```python
#Importamos las librerias principales
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/kidney_disease.csv")
```

```python
data
```

```python
data.drop('id', axis = 1, inplace = True)
```

```python
data
```

```python
data.info()
```

```python
data.classification.unique()
```

```python
data.classification=data.classification.replace("ckd\t","ckd")
```

```python
data.classification.unique()
```

```python
data.head()
```

```python
data['classification'] = data['classification'].replace(['ckd','notckd'], [1,0])
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
df = data.dropna(axis = 0)
print(f"Before dropping all NaN values: {data.shape}")
print(f"After dropping all NaN values: {df.shape}")
```

```python
df.head()
```

```python
df.index = range(0,len(df),1)
df.head()
```

```python
for i in df['wc']:
    print(i)
```

```python
df['wc']=df['wc'].replace(["\t6200","\t8400"],[6200,8400])
```

```python
for i in df['wc']:
    print(i)
```

```python
df.info()
```

```python
df['pcv']=df['pcv'].astype(int)
df['wc']=df['wc'].astype(int)
df['rc']=df['rc'].astype(float)
df.info()
```

```python
object_dtypes = df.select_dtypes(include = 'object')
object_dtypes.head()
```

```python
dictonary = {
        "rbc": {
        "abnormal":1,
        "normal": 0,
    },
        "pc":{
        "abnormal":1,
        "normal": 0,
    },
        "pcc":{
        "present":1,
        "notpresent":0,
    },
        "ba":{
        "notpresent":0,
        "present": 1,
    },
        "htn":{
        "yes":1,
        "no": 0,
    },
        "dm":{
        "yes":1,
        "no":0,
    },
        "cad":{
        "yes":1,
        "no": 0,
    },
        "appet":{
        "good":1,
        "poor": 0,
    },
        "pe":{
        "yes":1,
        "no":0,
    },
        "ane":{
        "yes":1,
        "no":0,
    }
}
```

```python
df=df.replace(dictonary)
```
